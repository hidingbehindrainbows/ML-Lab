```python
import pandas as pd

from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("/content/Iris.csv")

df.head()
```

```python
X = df.drop("Species",axis=1)
y = df["Species"]
```

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(data=df, x=df["SepalLengthCm"], y=df["SepalWidthCm"], legend=False, hue=df["Species"])

# Add labels and title
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Scatter Plot ofSepal Length vs Sepal Width')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(data=df, x=df["PetalLengthCm"], y=df["SepalWidthCm"], legend=False, hue=df["Species"])

# Add labels and title
plt.xlabel('Petal Length')
plt.ylabel('Sepal Width')
plt.title('Scatter Plot of Petal Length vs Sepal Width')

# Show the plot
plt.show()
```
