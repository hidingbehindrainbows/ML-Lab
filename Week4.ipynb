```python
# Import the libraries
from sklearn.datasets import make_classification
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
import numpy as np
```

```python
# Load the dataset
iris = load_iris()
X = iris.data[:, 2:] # petal length and width
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True, random_state=1)
```

```python
# DecisionTreeClassifier
model = tree.DecisionTreeClassifier()
model = model.fit(X_train, y_train)
```

```python
predicted_value = model.predict(X_test)
print(f"Predicted Value by the Model: {predicted_value}")
```

```python
# Plot the decision tree graph
tree.plot_tree(model)

zeroes = np.sum(y_train == 0)
ones = np.count_nonzero(y_train)

val = 1 - ((zeroes/70)**2 + (ones/70)**2)
print(f"Number of Zeros: {zeroes}\nNumber of Non-Zeroes: {ones}\nGini : {val}")
```

```python
#Print the Accuracy
match = sum([1 for i in range(30) if predicted_value[i] == y_test[i]])

accuracy = match/30
print(f"Accuracy is: {accuracy}")
```
